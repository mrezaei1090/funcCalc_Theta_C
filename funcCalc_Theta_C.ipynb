```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcCalc_Theta_C (x , y):
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    sig_xy_p = 0
    sig_xp = 0
    for i in range(len(x)):
        xp = np.around((x[i] - mean_x),4)
        yp = np.around((y[i] - mean_y),4)
        sig_xy_p = sig_xy_p + (xp * yp)
        sig_xp = sig_xp + xp ** 2
    theta = np.around((sig_xy_p / sig_xp),4)
    C = np.around((mean_y - (theta * mean_x)),4)
    return  theta , C
 
```

```python
a = np.random.randint(0,10,5)
b = np.random.randint(0,10,5)

print('a : ', a)
print('b : ', b)
print('theta , C  : ', funcCalc_Theta_C (a,b)) 
```

```python
plt.plot(a,b)
plt.show()
```
